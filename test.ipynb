```python
import sys
sys.path.append('../')

import src.countries as C
import matplotlib.pyplot as plt
import wbgapi as wb
```

```python
metrics = {
  'gdp_cap': 'NY.GDP.PCAP.CD',
  'edu_exp': 'SE.XPD.TOTL.GD.ZS',
  'elec_access': 'EG.ELC.ACCS.ZS',
  'power_cons': 'EG.USE.ELEC.KH.PC',
  'gdp_unitenergy': 'EG.GDP.PUSE.KO.PP.KD',
  'female_labor': 'SL.TLF.TOTL.FE.ZS',
  'fem_life_exp': 'SP.DYN.LE00.FE.IN',
}

countries = [C.India, C.USA, C.China, C.Japan, C.Germany]

df = wb.data.DataFrame(
  [metrics['fem_life_exp']],
  countries,
  [i*4+2003 for i in range(7)],
).transpose()
# df = df.pct_change(axis=0).round(2)

df = df.round(2)

df
```

```python
for country in countries:
  plt.plot(df[country.code], label=country.name)

plt.title('Female Life at Birth (2003-2023)')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.grid()
plt.show()
```
